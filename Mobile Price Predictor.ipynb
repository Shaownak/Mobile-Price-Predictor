```python
from sklearn.ensemble import RandomForestRegressor
import pandas as pd
```

```python
# Reading Data Set
training_data = pd.read_csv('train (1).csv')
training_data.describe()
```

```python
# Selecting Target
y = training_data.price_range
```

```python
# features
features = ['battery_power', 'clock_speed', 'dual_sim', 'four_g', 'int_memory', 'n_cores', 'px_height', 'px_width', 'ram',
           'three_g', 'touch_screen']
X = training_data[features]
```

```python
# Select model
model = RandomForestRegressor(random_state=1)
model.fit(X, y)
```

```python
predicted_price = model.predict(X)
```

```python
# Read test data
test_data = pd.read_csv('test.csv')
val_X = test_data[features]
```

```python
# Predict test data
predicted_test_data = model.predict(val_X)
predicted_test_data
```
